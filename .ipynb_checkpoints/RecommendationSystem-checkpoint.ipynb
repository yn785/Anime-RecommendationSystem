```python
# IF YOU HAVE ALL THE DATA ALREADY START FROM HERE 'Reviews.json'

def SimilarityMatrix(data):
    import time
    import warnings
    warnings.filterwarnings('ignore')
    import pandas as pd
    matrix = pd.DataFrame()
    starttime=time.time()
    for column in data:
        mean = data[data[column] != 0][column].mean()
        matrix[column] = data[column].replace(0,mean)
    print(time.time()-starttime)
    matrix = matrix.astype('int')
    # time by replacing data everytime 227.80549430847168 time 
    # time by creating a new dataframe and adding new improved columns 95.29253387451172
    print(matrix)
# Need to check the value type cz it's not giving the mean properly
def Cleaning(Filename):
    #Creating a pandas dataframe
    import pandas as pd
    import json
    f = open(Filename)
    data1 = json.load(f)
    data = pd.read_json(data1)
    #Cleaning the data
    #checking if there is any 0 value in data
    data.eq(0).any().any() 
    #changing "-" value to null
    import numpy as np
    data = data.mask(data == '')
    data = data.replace("-", pd.NA)
    data = data.replace({np.nan: pd.NA})
    # dropping all users with no scoring 
    data = data.dropna(how='all')
    #dropping all anime with no ratings
    data = data.dropna(how='all', axis=1)
    data = data.replace(pd.NA, 0)
    data = data.astype('int')
    SimilarityMatrix(data)
Cleaning('Reviews.json')
```
